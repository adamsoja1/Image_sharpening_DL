```python
import numpy as np
import os
import matplotlib.pyplot as plt
import cv2

def image_load_generator_x(path,batch_size):
    files = os.listdir(f'{path}')
    L = len(files)
    while True:
        batch_start = 0
        batch_size_end = batch_size
        while batch_start < L:
            limit = min(batch_size_end,L)
            
            files_batched = files[batch_start:limit]
            
            #loading data
            x_train = []
            
            for file in files_batched:
                X_train = plt.imread(f'{path}/{file}')
            
                image = cv2.resize(X_train, (230, 230)) 
                x_train.append(image)

            

        
            yield(np.array(x_train))
            
            batch_start +=batch_size
            batch_size_end +=batch_size
            
            
            
def image_load_generator_blurry(path,batch_size):

    files = os.listdir(f'{path}')
    L = len(files)
    while True:
        batch_start = 0
        batch_size_end = batch_size
        while batch_start < L:
            limit = min(batch_size_end,L)
            
            files_batched = files[batch_start:limit]
            
            #loading data
            x_train = []
            
            for file in files_batched:
                X_train = plt.imread(f'{path}/{file}')
                
                image = cv2.resize(X_train, (230, 230)) 
                image = cv2.GaussianBlur(image, (5, 5), cv2.BORDER_DEFAULT)
                
  
                x_train.append(image)

            

        
            yield(np.array(x_train))
            
            batch_start +=batch_size
            batch_size_end +=batch_size
            
```

```python
batch_size = 10
datagen = image_load_generator_x('flickr30k_images/flickr30k_images',batch_size)
datagen_blrr = image_load_generator_blurry('flickr30k_images/flickr30k_images',batch_size)
```

```python
X = next(datagen)
Y = next(datagen_blrr)
```

```python
import random
number = random.randint(0,batch_size-1)
plt.imshow(X[number])
plt.show()
plt.imshow(Y[number])
plt.show()
```
