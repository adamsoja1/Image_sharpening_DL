```python
import numpy as np
import matplotlib.pyplot as plt
import os
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import cv2 
```

```python

path = 'flickr30k_images'
files = os.listdir(path)
```

```python
files = files[-40:len(files)]
```

```python
for i in range(5):
    image = plt.imread(f'{path}/{files[i]}')
    plt.imshow(image)
    plt.show()
```

```python
def add_noise(img):
    image = cv2.GaussianBlur(img, (3, 3), cv2.BORDER_DEFAULT)

    return image




args = dict(
            preprocessing_function=add_noise,
           )
```

```python
img_datagen = ImageDataGenerator(**args)
```

```python
images  = img_datagen.flow_from_directory(path,
                                          target_size=(250,250),
                                batch_size=5)

```

```python
xd,xd2 = next(images)
xd.shape
xd = (xd/xd.max()) * 255
```

```python
for i in range(5):
    plt.imshow((xd[i]).astype(np.uint8))
    plt.show()
    
```
